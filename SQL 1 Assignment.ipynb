```python
q = """SELECT
            sex
        , SUM(hours_per_week) as average_hours_per_week
         FROM
              sqladb
           GROUP BY
              
              hours_per_week;"""
```

```python
(pysqldf(q)) # average hours per week of all men who are working in private sector
```

```python
# age distribution by country
```

```python
sqladb.native_country.replace("?",np.nan).dropna(inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
sqladb['native_country']=le.fit_transform(sqladb.native_country.values)
```

```python
q1 = """SELECT
            Age
        , SUM(native_country) as Country
         FROM
              sqladb
           GROUP BY
              
              native_country;"""
```

```python
pysqldf(q1) # age distribution by country
```
